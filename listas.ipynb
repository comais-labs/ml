### List Comprehensions

```python
[f.upper() for f in frutas]
```

```python
lista = [2,3,4,5,6,7]
```

```python
lista_quadrado = [x**2 for x in lista]
lista_quadrado
```

```python
[x for x in lista_quadrado if x%2 ==0]
```
