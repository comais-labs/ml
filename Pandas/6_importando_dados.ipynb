```python
import pandas as pd
```

```python
df = pd.read_excel('dados/dados.xlsx')
```

```python
df_unimed = pd.read_html('https://www.unimed.coop.br/viver-bem/pais-e-filhos/estatura-por-idade', encoding='ISO-8859-1' )
```

```python
df_menina_02 = df_unimed[0]
```

```python
df_menina_02.rename(columns={0:"Idade",1:"Minimo",2:"Media", 3:"Maximo" }, inplace=True)
```

```python
df_menina_02.drop(0, inplace=True)
```

```python
df_menina_02
```

```python
def idade_int(x):
    x = x.split(' ')[0]
    x = int(x)
    return x
    
```

```python
df_menina_02['Idade'] = df_menina_02['Idade'].apply(idade_int)
```

```python
df_menina_02
```

```python
def converte_num(x):
    x = x.split(' ')[0]
    x = x.replace(",",".")
    x = float(x)
    return x
```

```python
df_menina_02['Minimo'] = df_menina_02['Minimo'].apply(converte_num)
```

```python
df_menina_02
```

```python
df_menina_02['Media'] = df_menina_02['Media'].apply(converte_num)
df_menina_02['Maximo'] = df_menina_02['Maximo'].apply(converte_num)
```

```python
df_menina_02.info()
```

```python
df_menina_02.describe()
```

```python
df_menina_02.to_csv('dados/meninas_02Anos.csv')
```

```python
df_menina_02.to_excel('dados/meninas_02Anos.xlsx')
```

```python
df_menina_02.to_parquet('dados/menias_02Anos.pqt')
```
