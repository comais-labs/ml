### 1. Introdução à Normalização

Normalização de dados é o processo de redimensionar os atributos para que tenham uma escala comum. Isso é importante em algoritmos de machine learning que dependem da distância entre os pontos de dados, como K-Means e SVM, pois garante que uma característica não domine as outras devido à sua escala.

### 2. Preparação do Ambiente

Comece importando as bibliotecas necessárias para manipulação de dados e normalização:

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
```

### 3. Criando um Conjunto de Dados de Exemplo

Vamos criar um DataFrame de exemplo com dados numéricos que representam idades e salários.

```python
data = {
    'Idade': [25, 30, 35, 40, 45],
    'Salário': [4000,50000, 60000, 80000, 90000],
    'Cargo': ['Engenheiro', 'Cientista', 'Engenheiro', 'Cientista', 'Engenheiro'],

}

df = pd.DataFrame(data)
df
```

### 4. Normalização Min-Max

A normalização Min-Max redimensiona os dados para um intervalo específico, geralmente 0 a 1, sem distorcer as diferenças nos intervalos de valores ou perder informações.


```python
# Aplicando a normalização Min-Max
scaler_min_max = MinMaxScaler()
df_min_max_scaled = pd.DataFrame(scaler_min_max.fit_transform(df[['Idade','Salário']]), columns=['Idade','Salário'])

# Visualizando os dados normalizados com Min-Max
print("\nDados Normalizados com Min-Max:")
print(df_min_max_scaled)
```

### 5. Normalização Z-Score (Padronização)

A padronização (ou normalização Z-score) redimensiona os dados para que tenham média (µ) 0 e desvio padrão (σ) 1.

```python
# Aplicando a padronização Z-Score
scaler_z_score = StandardScaler()
df_z_score_scaled = pd.DataFrame(scaler_z_score.fit_transform(df[['Idade','Salário']]), columns=[['Idade','Salário']])

# Visualizando os dados padronizados (Z-Score)
print("\nDados Padronizados (Z-Score):")
print(df_z_score_scaled)
```

### 6. Visualização dos Dados

A visualização ajuda a entender o efeito da normalização e da padronização nos dados.

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Dados originais
axs[0].scatter(df['Idade'], df['Salário'])
axs[0].set_title('Dados Originais')
axs[0].set_xlabel('Idade')
axs[0].set_ylabel('Salário')
#Dados normalizados com Min-Max
axs[1].scatter(df_min_max_scaled['Idade'], df_min_max_scaled['Salário'])
axs[1].set_title('Dados Normalizados com Min-Max')
axs[1].set_xlabel('Idade')
axs[1].set_ylabel('Salário')
# Dados padronizados (Z-Score)
axs[2].scatter(df_z_score_scaled['Idade'], df_z_score_scaled['Salário'])
axs[2].set_title('Dados Padronizados (Z-Score)')
axs[2].set_xlabel('Idade')
axs[2].set_ylabel('Salário')
```

```python
#Dados normalizados com Min-Max
axs[1].scatter(df_min_max_scaled['Idade'], df_min_max_scaled['Salário'])
axs[1].set_title('Dados Normalizados com Min-Max')
axs[1].set_xlabel('Idade')
axs[1].set_ylabel('Salário')
```

```python
# Dados padronizados (Z-Score)
axs[2].scatter(df_z_score_scaled['Idade'], df_z_score_scaled['Salário'])
axs[2].set_title('Dados Padronizados (Z-Score)')
axs[2].set_xlabel('Idade')
axs[2].set_ylabel('Salário')
```

```python
plt.tight_layout()
plt.show()
```

A normalização e a padronização são técnicas cruciais de pré-processamento em Machine Learning. A escolha entre elas depende do algoritmo específico que você está utilizando e do conjunto de dados em questão. A normalização Min-Max é útil quando precisamos de valores positivos em uma escala específica, enquanto a padronização é benéfica em casos que requerem dados com média zero, útil especialmente para algoritmos que assumem que os dados estão normalmente distribuídos. Ambas as técnicas ajudam a melhorar o desempenho dos modelos de machine learning ao garantir que as características contribuam igualmente para a aprendizagem do modelo.
