Agrupando dados por Região e somando o Total:

```python
def estado_para_regiao(sigla_estado):
    # Dicionário com as regiões e suas respectivas siglas de estados
    estados_brasil = {
        "Norte": ["AC", "AP", "AM", "PA", "RO", "RR", "TO"],
        "Nordeste": ["AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"],
        "Centro-Oeste": ["GO", "MT", "MS", "DF"],
        "Sudeste": ["ES", "MG", "RJ", "SP"],
        "Sul": ["PR", "RS", "SC"]
    }

    if sigla_estado in estados_brasil['Norte']: 
        return "Norte"
    if sigla_estado in estados_brasil['Nordeste']: 
        return "Nordeste"
    if sigla_estado in estados_brasil['Centro-Oeste']: 
        return "Centro-Oeste"
    if sigla_estado in estados_brasil['Sudeste']: 
        return "Sudeste"
    if sigla_estado in estados_brasil['Sul']: 
        return "Sul"
#Os ifs acima podem ser subistituidos pelo codigo abaixo
#    for regiao, siglas in estados_brasil.items():
#        if sigla_estado.upper() in siglas:
#            return regiao
    
    return "Sigla de estado inválida"


```

```python
estado_para_regiao('TO')
```

```python
df['UF'].apply(estado_para_regiao)
```

```python
df.insert(loc=3, column="REGIÃO",value = df["UF"].apply(estado_para_regiao))
```
